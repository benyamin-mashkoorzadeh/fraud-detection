```python
import nbformat
import time
import os
import json
import sys
from IPython.core.interactiveshell import InteractiveShell
from IPython.display import clear_output


notebook_path = "/Users/benyamin/Desktop/Dev/Projects/Laravel/tus/backend/public/python_scripts/program.ipynb"

# ۱. تعریف متغیر `file_path`
# file_path = "/Users/benyamin/Desktop/Thesis-File/creditcard.csv"

# مسیر ذخیره فایل در دایرکتوری storage لاراول
progress_file = "/Users/benyamin/Desktop/Dev/Projects/Laravel/tus/backend/storage/app/progress.json"





def save_progress(progress, status="running", message="Processing..."):
    # چاپ داده‌ها برای بررسی
    print(f"Saving progress: {progress}, Status: {status}, Message: {message}")
    
    # ذخیره به فایل JSON
    with open(progress_file, "w") as f:
        json.dump({
            "progress": progress, 
            "status": status, 
            "message": message
        }, f)
    
    # بررسی محتوای فایل بعد از ذخیره‌سازی
    with open(progress_file, "r") as f:
        data = json.load(f)
        print("File content:", data)

        
# دریافت مسیر فایل از متغیر محیطی
file_path = os.environ.get("FILE_PATH", "")

print(f"Received file path: {repr(file_path)}")  # لاگ برای بررسی

# بررسی اینکه مسیر فایل دریافت شده یا نه
if not file_path:
    save_progress(0, status="error", message="No file path received")
    print("Error: No file path received")
    sys.exit(1)

# بررسی اینکه فایل وجود دارد یا نه
if not os.path.exists(file_path):
    save_progress(0, status="error", message="File not found")
    print(f"Error: File not found - {file_path}")
    sys.exit(1)

# پردازش داده‌ها
# def process_data(file_path):
#     print(f"Processing file: {file_path}")
    
#     total_steps = 100
#     for step in range(1, total_steps + 1):
#         time.sleep(1)  # شبیه‌سازی زمان پردازش
#         progress = int((step / total_steps) * 100)
#         save_progress(progress, status="running", message=f"Processing... {progress}%", file_path=file_path)  # ذخیره درصد پیشرفت
#         print(f"Progress: {progress}%")  # نمایش در لاگ


# ذخیره وضعیت شروع پردازش
save_progress(0, message="Starting process...")

# اجرای پردازش
# process_data(file_path)









# ۲. خواندن نوت‌بوک و شمارش تعداد سلول‌های کد
with open(notebook_path, "r", encoding="utf-8") as f:
    notebook = nbformat.read(f, as_version=4)

code_cells = [cell for cell in notebook.cells if cell.cell_type == "code"]
total_cells = len(code_cells)

# ۳. ساخت یک اینترپرتر برای اجرای کدها
shell = InteractiveShell.instance()

# ۴. انتقال متغیر `file_path` به محیط اجرایی
shell.run_cell(f"file_path = '{file_path}'")

# ۵. اجرای سلول‌ها یکی‌یکی و نمایش درصد پیشرفت
for i, cell in enumerate(code_cells):
    shell.run_cell(cell.source)  # اجرای کد داخل هر سلول
    
    # محاسبه درصد پیشرفت
    percent = ((i + 1) / total_cells) * 100
    
    # نمایش درصد پیشرفت به‌صورت عددی
    clear_output(wait=True)  # پاک کردن خروجی قبلی
    # print(f"********Progress***********: {percent:.2f}%")
    save_progress(f"{percent:.0f}", status="success", message=f"Processing... {percent:.0f}%")
    time.sleep(0.2)  # شبیه‌سازی تأخیر کوچک

# بعد از تکمیل پردازش
save_progress(100, status="success", message="Execution completed successfully")
print("Success: Execution completed successfully!")


```
